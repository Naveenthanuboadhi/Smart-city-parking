```python
from sklearn.impute import SimpleImputer
numerical_cols = ['TOTALSPACECOUNT', 'EVSPACECOUNT', 'ADASPACECOUNT', 'RESERVEDSPACECOUNT', 'NUMBEROFLEVELS', 'BIPAREA']
numerical_imputer = SimpleImputer(strategy='median')
df[numerical_cols] = numerical_imputer.fit_transform(df[numerical_cols])
```

```python
# Imputing categorical columns
categorical_cols = ['SALESFORCEID', 'SUBTYPEID', 'VALIDATION', 'PARKINGTYPE', 'PERMITS', 'BIPLEASE']
categorical_imputer = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = categorical_imputer.fit_transform(df[categorical_cols])
```

```python
df['CREATIONDATE'] = df['CREATIONDATE'].fillna(method='ffill')
```
